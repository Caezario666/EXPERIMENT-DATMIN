```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'C:/Semester 5/Data Mining/Coba/jumlah-fasilitas-kesehatan-di-kabupaten-semarang-tahun-2023.xlsx'
data = pd.read_excel(file_path, sheet_name='Sheet1')
```

```python
print("Data Sample:")
print(data.head())
```

```python
data.columns = ['id', 'uraian', 'satuan', 'jumlah_2023']
data = data.dropna(subset=['jumlah_2023'])
```

```python
print("\nJumlah fasilitas kesehatan per jenis pada tahun 2023:")
facility_counts = data.groupby('uraian')['jumlah_2023'].sum()
print(facility_counts)
```

```python
plt.figure(figsize=(10, 6))
facility_counts.plot(kind='bar', color='skyblue')
plt.title('Jumlah Fasilitas Kesehatan di Kabupaten Semarang Tahun 2023')
plt.xlabel('Jenis Fasilitas')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
rawat_inap = data[data['uraian'] == 'a. Rawat Inap']['jumlah_2023'].values[0]
rawat_jalan = data[data['uraian'] == 'b. Rawat Jalan']['jumlah_2023'].values[0]
```

```python
print(f"\nJumlah Puskesmas Rawat Inap: {rawat_inap}")
print(f"Jumlah Puskesmas Rawat Jalan: {rawat_jalan}")
```

```python
labels = ['Rawat Inap', 'Rawat Jalan']
values = [rawat_inap, rawat_jalan]
```

```python
plt.figure(figsize=(6, 4))
plt.bar(labels, values, color=['salmon', 'lightgreen'])
plt.title('Perbandingan Puskesmas Rawat Inap vs Rawat Jalan')
plt.xlabel('Jenis Layanan')
plt.ylabel('Jumlah')
plt.show()
```

```python
print("\nKesimpulan:")
if rawat_inap < rawat_jalan:
    print("Jumlah Puskesmas dengan layanan Rawat Jalan lebih banyak dibandingkan dengan layanan Rawat Inap.")
else:
    print("Jumlah Puskesmas dengan layanan Rawat Inap lebih banyak atau sama dengan layanan Rawat Jalan.")
```
